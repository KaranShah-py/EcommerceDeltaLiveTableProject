#Customer Order Summary

## The following notebook creates gold table 
- Join the tables Silver_customers, Silver_Addresses, Silver_orders
- Retrieve the latest address of customers, 
- Calculate the following values 
  1. total_orders, 
  2. total_items_ordered
  3. total_order_amount 

```python
%sql
CREATE OR REFRESH MATERIALIZED VIEW ecommercedataset.goldlayer.goldCustomerOrderSummary
AS 
  SELECT
    C.customer_id, 
    C.customer_name, 
    C.date_of_birth, 
    C.telephone, 
    C.email, 
    A.address_line_1,
    COUNT(DISTINCT order_id) AS COUNT_OF_ORDERS, 
    SUM(O.item_quantity) AS total_items_ordered,
    SUM(O.item_quantity * O.item_price) AS total_order_value
  FROM LIVE.ecommercedataset.silverlayer.CustomerSilverSCD AS C
  JOIN LIVE.ecommercedataset.silverlayer.AddressesSilver AS A 
  ON C.customer_id = A.customer_id 
  JOIN LIVE.silver_orders O
  ON C.customer_id = O.customer_id 
  WHERE 
    a.__END_AT IS NULL -- latest address records;
  GROUP BY ALL;
```
